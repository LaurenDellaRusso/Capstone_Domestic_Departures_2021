#### ONLY KEEP DEPARTURE ROWS THAT ARE PART OF LARGE HUB AIRPORTS
#### LARGE HUB = DEPARTURES Departures from all Large hub airports to any destination

```python
large_hubs_list = ['ATL','DFW','DEN','ORD','LAX','CLT','MCO','LAS','PHX','MIA','SEA','IAH','JFK','EWR','FLL','MSP',
                        'SFO','DTW','BOS','SLC','PHL','BWI','TPA','SAN','LGA','MDW','BNA','IAD']
```

```python
large_hubs_2021 = all2021all.loc[all2021all['departure'].isin(large_hubs_list)].reset_index(drop=True)
large_hubs_2021
```

```python
#change days of week from numeric to actual day

for index, row in large_hubs_2021.iterrows():

    if large_hubs_2021.loc[index, 'DayOfWeek'] == 1:
        large_hubs_2021.loc[index, 'DayOfWeek'] = 'Monday'
    elif large_hubs_2021.loc[index, 'DayOfWeek'] == 2:
        large_hubs_2021.loc[index, 'DayOfWeek'] = 'Tuesday'
    elif large_hubs_2021.loc[index, 'DayOfWeek'] == 3:
        large_hubs_2021.loc[index, 'DayOfWeek'] = 'Wednesday'
    elif large_hubs_2021.loc[index, 'DayOfWeek'] == 4:
        large_hubs_2021.loc[index, 'DayOfWeek'] = 'Thursday'
    elif large_hubs_2021.loc[index, 'DayOfWeek'] == 5:
        large_hubs_2021.loc[index, 'DayOfWeek'] = 'Friday'
    elif large_hubs_2021.loc[index, 'DayOfWeek'] == 6:
        large_hubs_2021.loc[index, 'DayOfWeek'] = 'Saturday'
    elif large_hubs_2021.loc[index, 'DayOfWeek'] == 7:
        large_hubs_2021.loc[index, 'DayOfWeek'] = 'Sunday'
    elif large_hubs_2021.loc[index, 'DayOfWeek'] == 9:
        large_hubs_2021.loc[index, 'DayOfWeek'] = 'Unknown'
```

```python
#add a column to flight times of day

for index, row in large_hubs_2021.iterrows():
    
    if large_hubs_2021.loc[index, 'DepTime'] < 559:
        large_hubs_2021.loc[index, 'time_of_day'] = 'Overnight midnight-6am'
    if large_hubs_2021.loc[index, 'DepTime'] >= 600:
        large_hubs_2021.loc[index, 'time_of_day'] = 'Early 6-9am'
    if large_hubs_2021.loc[index, 'DepTime'] >= 901:
        large_hubs_2021.loc[index, 'time_of_day'] = 'Morning 9am-noon'
    if large_hubs_2021.loc[index, 'DepTime'] >= 1201:
        large_hubs_2021.loc[index, 'time_of_day'] = 'Afternoon noon-3pm'
    if large_hubs_2021.loc[index, 'DepTime'] >= 1501:
        large_hubs_2021.loc[index, 'time_of_day'] = 'Late Afternoon 3-6pm'
    if large_hubs_2021.loc[index, 'DepTime'] >= 1801:
        large_hubs_2021.loc[index, 'time_of_day'] = 'Evening 6-9pm'
    if large_hubs_2021.loc[index, 'DepTime'] >= 2101:
        large_hubs_2021.loc[index, 'time_of_day'] = 'Late 9pm-midnight'
```

```python
large_hub_2021_clean = large_hubs_2021[(large_hubs_2021['arrival'].notnull()) & (large_hubs_2021['airline_code'].notnull()) & (large_hubs_2021['airline'].notnull()) & (large_hubs_2021['departure'].notnull())]
print('Number of Rows: ' + str(len(large_hub_2021_clean)))

#data is clean - no rows removed
```
