```python
import requests
import matplotlib.pyplot as plt
import json
import pandas as pd
import numpy as np
import time 


%matplotlib inline
```

```python
from amadeus import Client, Location

amadeus = Client(
    client_id='',
    client_secret=''
)
```

```python
airport_list = ['ATL','DFW','DEN','ORD','LAX','CLT','MCO','LAS','PHX','MIA','SEA','IAH','JFK','EWR','FLL','MSP',
                'SFO','DTW','BOS','SLC','PHL','BWI','TPA','SAN','LGA','MDW','BNA','IAD']
```

```python
detailedname_list = []
#iatacode_list = []
latitude_list = []
longitude_list = []
#score_list = []

for code in airport_list:
    response = amadeus.reference_data.locations.get(
        keyword=code,
        subType=Location.AIRPORT
    )
    name = response.data[0]['detailedName']
    #iatacode = response.data[0]['iataCode'] 
    lat = response.data[0]['geoCode']['latitude']
    lon = response.data[0]['geoCode']['longitude']
    #score = response.data[0]['analytics']['travelers']['score']
 
    
    detailedname_list.append(name)
    #iatacode_list.append(iatacode)
    latitude_list.append(lat)
    longitude_list.append(lon)
    #score_list.append(score)

    
    time.sleep(1)
```

```python
airports_dict = {
    'name': detailedname_list,
    'departure':airport_list,
    #'iataCode':iatacode_list,
    'latitude':latitude_list,
    'longitude':longitude_list#,
    #'score':score_list
}

airport = pd.DataFrame(airports_dict)
airport
```

```python
delaysandlocs = pd.merge(airport, delays, how='left', left_on='departure', right_on='departure')
delaysandlocs
```
