# Departure delays by airport vs. BNA

##### Total Departing Flights by Large Hub

```python
#total departing flights by hub
total_flights= main2021.groupby(["departure"])["departure"].count().reset_index(name="total_flights")
total_flights.sort_values('total_flights', ascending = False).reset_index(drop=True)
```

##### Calculate Delayed Departing Flights by Large Hub

```python
#total delayed flights by hub
total_delays = delaysdf.groupby(["departure"])["departure"].count().reset_index(name="total_delays")
```

##### Calculate % of Delayed Departing Flights by Large Hub

```python
hubdelays = pd.merge(total_flights, total_delays, on='departure', how='outer')
hubdelays['percent_delay'] = pd.DataFrame((hubdelays['total_delays']/hubdelays['total_flights']*100).round(2))
hubdelays.sort_values('percent_delay', ascending = False).reset_index(drop=True)
```

```python
#hubdelays.to_csv('hubdelays.csv')
```
